<b>Displaying the details of matches held in Spain using different tables</b> 

```python
matches=pd.read_sql(""" select match.id,
                                country.name as country_name,
                                league.name as league_name,
                                season,
                                stage,
                                date,
                                HT.team_long_name as home_team,
                                AT.team_long_name as away_team,
                                home_team_goal,
                                away_team_goal
                        from    match join country on country.id=match.country_id
                                    join league on league.id=match.league_id
                                    left join team as HT on HT.team_api_id=match.home_team_api_id
                                    left join team as AT on AT.team_api_id=match.away_team_api_id
                        where   country_name='Spain'
                        order by date
                        limit 10;""",conn)
matches
```

<H2>Analysis of Data</H2>

<b>Displaying leagues by season for each country</b>

```python

leages_by_season = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
leages_by_season
```

<b>Visualising the average goals per game</b>

```python
leages_by_season.pivot_table(index=['season'],values=['avg_goals'],columns=['country_name']).plot(figsize=(12,5), 
                            title='Average Goals per Game over Time')
plt.legend(['Spain', 'Germany', 'France', 'Italy', 'England'],loc='best')
```

<b>Visualising the average goal difference</b>

```python
leages_by_season.pivot_table(index=['season'] ,values=['avg_goal_dif'], columns=['country_name']).plot(figsize=(12,5),
                            title='Average Goals Difference Home vs Away')
plt.legend(['Spain', 'Germany', 'France', 'Italy', 'England'],loc='best')
```

<b>Display players heights using subquery</b> 

```python
players_height = pd.read_sql("""SELECT CASE
                                        WHEN ROUND(height)<165 then 165
                                        WHEN ROUND(height)>195 then 195
                                        ELSE ROUND(height)
                                        END AS calc_height, 
                                        COUNT(height) AS distribution, 
                                        (avg(PA_Grouped.avg_overall_rating)) AS avg_overall_rating,
                                        (avg(PA_Grouped.avg_potential)) AS avg_potential,
                                        AVG(weight) AS avg_weight 
                            FROM PLAYER
                            LEFT JOIN (SELECT Player_Attributes.player_api_id, avg(Player_Attributes.overall_rating) AS avg_overall_rating,
                                        avg(Player_Attributes.potential) AS avg_potential  
                                        FROM Player_Attributes
                                        GROUP BY Player_Attributes.player_api_id) 
                                        AS PA_Grouped ON PLAYER.player_api_id = PA_Grouped.player_api_id
                            GROUP BY calc_height
                            ORDER BY calc_height
                                ;""", conn)
players_height
```

<b>Visualising the players height</b>

```python
players_height.plot(xlabel=['calc_height'],y=['avg_overall_rating'],figsize=(12,5),title='Potential vs Height')
```
